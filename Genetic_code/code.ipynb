```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
# img_path = 'images.jpg'
# img_path = 'images2.jpg'
# img_path = 'circle.png'
# img_path = 'circle2.png'
# img_path = 'circle3.png'
# img_path = 'circle4.png'
# img_path = 'circle5.png'
img_path = 'Ronaldo_WC22.webp'
# img_path = 'egg.webp'
img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import MyGenetic
history = MyGenetic.GA(edges, MaxIt = 40, nPop = 100, Ns = 500)
```

```python
import matplotlib.pyplot as plt

plt.plot(history["average_cost"], label = "AverageCosts")
plt.plot([ch.cost for ch in history['best_sol']], label = "BestCosts")

plt.title('GA Analyse')
plt.xlabel('Cost')
plt.ylabel('Iteration')

# show a legend on the plot
plt.legend()

# function to show the plot
plt.show()
```

```python
import matplotlib.image as mpimg
image = mpimg.imread(img_path)
plt.imshow(image)
plt.plot([x[1] for x in history["Population"][0].circle_points], [x[0] for x in history["Population"][0].circle_points])
plt.plot([x[1] for x in history["Population"][1].circle_points], [x[0] for x in history["Population"][1].circle_points])
plt.plot([x[1] for x in history["Population"][2].circle_points], [x[0] for x in history["Population"][2].circle_points])
plt.show()
```
